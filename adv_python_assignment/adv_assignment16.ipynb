```python
"""Q6. What are some examples of how to use the match object returned by re.match and re.search?
A =The re.search() and re.match() are function of python re module.
these function are very efficient and fast for searching the strings.Function searches string in substring if match is found then it return the pattern or else it retuns NONE
There is difference in use of both functions.Both retuns the pattern in substring found in string.but re.match() find only from begining from the string and return the first match object if found
but if string found somewhere in middle the it returns NONE"""
import re
u1 ="Ukrain is eastern european country near russia."
r1 = "Russia is eatern european country near ukrain,belarus,estonia,lativia and lituania"
sub ="Russia"
print(re.search(sub,r1,re.IGNORECASE))
print(re.match(sub,r1,re.IGNORECASE))
print(re.search(sub,u1,re.IGNORECASE))
print(re.match(sub,u1,re.IGNORECASE))
```
