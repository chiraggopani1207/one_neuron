```python
"""Q3. In a simple match of a string, which looks only for one match and does not do any replacement, is
the use of a nontagged group likely to make any practical difference?
A =in this case non tagged group will not make any significant difference."""
import re
phonenumber = re.compile(r'\d\d\d')
num = phonenumber.search("My phone number is --> 123-456-7890")
print(f'Phone number found -->',{num.group()})
print(f'Phone number found -->',{num.group(0)})
```
