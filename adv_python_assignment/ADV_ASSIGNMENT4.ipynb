```python
""""Q1. Which two operator overloading methods can you use in your classes to support iteration?
A = __iter__ and __next__ are two overloading operators used to iterate object and collectively called as iter tools
__iter__ return the iterator oject and called at the start of the loop at respective call
__next__ is called at each value of iter tool and stop iteration is used when no value is left
"""
class Counter:
    def __init__(self,low,high):
        self.current = low
        self.high = high
    def __iter__(self):
        return self
    def __next__(self):
        if self.current > self.high:
            raise StopIteration
        else:
            self.current += 1
            return self.current - 1
for ele in Counter(5,15):
    print(ele, end=" ")
        
```

```python
"""Q4. In a class, how do you capture in-place addition?
A =a+b is normal addition whereas a += b is inplace addition operation"""
class Book:
    def __init__(self,pages):
        self.pages = pages
    def __iadd__(self,other):
        self.pages += other.pages
        return self.pages
        
b1 = Book(100)
b2 = Book(200)
b1+=b2
print(b1)
```
