```python
"""1. Write a Python program to convert kilometers to miles?"""
def miles():
    """This function helps user to convert length of KM to Miles"""
    km =float(input("Enter length of Km you wish to convert into miles : "))
    return ("Distance converted from KM to MILES is equals to : " ,km * 0.621371)
```

```python
"""2. Write a Python program to convert Celsius to Fahrenheit?"""
def Fahrenheit():
    """This program helps user to convert temprature from celsius to Fahrenheit"""
    celsius = float(input("Enter celsius degree temprature to converst to equivalent to Fahrenheit : "))
    Fahrenheit = (celsius*(9/5) + 32)
    print("{} celsius is equal to {} Fahrenheit".format(celsius,Fahrenheit))
```

```python
Fahrenheit()
```

```python
"""5. Write a Python program to swap two variables without temp variable?"""
num1 = int(input("Enter number 1 : "))
num2 = int(input("Enter number 2 : "))

def swapnumber(num1,num2):
    """This function allows user to swap numbers of two variables withous using temp variable"""
    print("before swaping numbers 1 and 2 are : ", num1  , num2)
    num1 = num1+num2
    num2 = num1-num2
    num1 =num1-num2
    print("after swaping number 1 and 2 are : ", num1   , num2)
```
